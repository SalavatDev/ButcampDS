```python
img = cv2.imread('gb_zsqrK7.jpg', cv2.IMREAD_COLOR)
# img = cv2.imread('img.png', cv2.IMREAD_GRAYSCALE)
plt.imshow(img)
```

```python
rgb_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_img)
```

```python
# размытие по Гауссу
blurred = cv2.GaussianBlur(rgb_img, (51, 51), 0)
plt.imshow(blurred)
```

```python
x = cv2.Sobel(rgb_img, cv2.CV_16S,1,0)  
y = cv2.Sobel(rgb_img, cv2.CV_16S,0,1)  
  
absX = cv2.convertScaleAbs(x)
absY = cv2.convertScaleAbs(y) 

plt.imshow(absX)
```

```python
med1 = cv2.medianBlur(rgb_img, 15) # 5x5 медианный фильтр
plt.imshow(med1)
```
